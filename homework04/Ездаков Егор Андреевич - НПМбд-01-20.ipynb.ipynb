```python
import numpy
import math
def func5(x):
    return numpy.cos(x)**3
a, b = 0, math.pi/2
dx = 0.0001
def solve_integr(func, dx):
    n = int((b - a) / dx)
    x_step = np.linspace(a, b, n)
    meth1 = dx * np.sum(func(x_step))
    meth2 = dx / 2 * (func(x_step[0]) + 2 * np.sum(func(x_step[1:-1])) + func(x_step[-1]))
    meth3 = dx / 3 * (func(x_step[0]) + 2 * np.sum(func(x_step[2:-1:2])) + 4 * np.sum(func(x_step[1:-2:2])) + func(x_step[-1]))
    return meth1, meth2, meth3


print('Решение интеграла 8:', solve_integr(func5, dx)[0], ' (метод прямоугольников)')
print('Решение интеграла 8:', solve_integr(func5, dx)[1], ' (метод трапеций)')
print('Решение интеграла 8:', solve_integr(func5, dx)[2], ' (метод Симпсона)')
'''
При уменьшении dx уменьшается шаг и увеличенивается число точек x_step.Таким образом, увеличивается точность решения.
'''
```
